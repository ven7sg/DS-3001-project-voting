```python
import pandas as pd
```

```python
# Load historical voting data
vdf = pd.read_csv('./DS-3001-project-voting/data/voting_VA.csv')

# View historical voting data
vdf
```

```python
# Summary information of dataframe (Tamera Fang)
vdf.describe()
```

```python
# Remove unnecessary columns in vdf:
voting_cleaned = vdf.drop(columns=['Unnamed: 0', 'version', 'mode'])
unnecessary_columns = ['office', 'state', 'state_po'] # office bc all of them are US pres and state bc all are VA
voting_cleaned.drop(unnecessary_columns, axis=1, inplace=True)
voting_cleaned.dropna(inplace=True)
voting_cleaned['party'] = voting_cleaned['party'].astype('category')
voting_cleaned
```

```python
# Check candidate column for unique values (Anna)
voting_cleaned['candidate'].unique()

# Combine Donald Trump and Donald J Trump (same candidate)
voting_cleaned['candidate'].replace('DONALD J TRUMP', 'DONALD TRUMP', inplace=True)

# Recheck unique values
voting_cleaned['candidate'].unique()
```

```python
# Check county name column for unique values (Anna)
voting_cleaned['county_name'].unique()
```

```python
# Combine duplicates
voting_cleaned['county_name'].replace({
    'ALEXANDRIA': 'ALEXANDRIA CITY',
    'BRISTOL': 'BRISTOL CITY',
    'BUENA VISTA': 'BUENA VISTA CITY',
    'CharlesCty': 'CHARLES CITY',
    'CHARLOTTESVILLE': 'CHARLOTTESVILLE CITY',
    'CHESAPEAKE': 'CHESAPEAKE CITY',
    'COLONIAL HEIGHTS': 'COLONIAL HEIGHTS CITY',
    'COVINGTON': 'COVINGTON CITY',
    'DANVILLE': 'DANVILLE CITY',
    'EMPORIA': 'EMPORIA CITY',
    'FAIRFAX': 'FAIRFAX CITY',
    'FALLS CHURCH': 'FALLS CHURCH CITY',
    'FREDERICKSBURG': 'FREDERICKSBURG CITY',
    'HAMPTON': 'HAMPTON CITY',
    'HARRISONBURG': 'HARRISONBURG CITY',
    'LEXINGTON': 'LEXINGTON CITY',
    'LYNCHBURG': 'LYNCHBURG CITY',
    'MANASSAS': 'MANASSAS CITY',
    'MARTINSVILLE': 'MARTINSVILLE CITY',
    'NEWPORT NEWS': 'NEWPORT NEWS CITY',
    'NORFOLK': 'NORFOLK CITY',
    'NORTON': 'NORTON CITY',
    'PETERSBURG': 'PETERSBURG CITY',
    'POQUOSON': 'POQUOSON CITY',
    'PORTSMOUTH': 'PORTSMOUTH CITY',
    'RADFORD': 'RADFORD CITY',
    'SALEM': 'SALEM CITY',
    'STAUNTON': 'STAUNTON CITY',
    'SUFFOLK': 'SUFFOLK CITY',
    'VIRGINIA BEACH': 'VIRGINIA BEACH CITY',
    'WAYNESBORO': 'WAYNESBORO CITY',
    'WILLIAMSBURG': 'WILLIAMSBURG CITY',
    'WINCHESTER': 'WINCHESTER CITY'
}, inplace=True)

# Recheck check unique values
voting_cleaned['county_name'].unique()
```

```python
# Create an empty DataFrame to store the combined results (Anna)
voting_combined = pd.DataFrame(columns=voting_cleaned.columns)

# Iterate over unique combinations of all columns except 'candidatevotes'
unique_combinations = voting_cleaned.drop(columns=['candidatevotes']).drop_duplicates()

for idx, row in unique_combinations.iterrows():
    # Filter the original DataFrame to get rows with identical values except 'candidatevotes'
    filtered_rows = voting_cleaned[
        (voting_cleaned['year'] == row['year']) &
        (voting_cleaned['county_name'] == row['county_name']) &
        (voting_cleaned['county_fips'] == row['county_fips']) &
        (voting_cleaned['candidate'] == row['candidate']) &
        (voting_cleaned['party'] == row['party']) &
        (voting_cleaned['totalvotes'] == row['totalvotes'])
    ]

    # Combine the 'candidatevotes' values
    combined_votes = filtered_rows['candidatevotes'].sum()

    # Add the combined row to the result DataFrame
    combined_row = row.copy()
    combined_row['candidatevotes'] = combined_votes
    voting_combined = pd.concat([voting_combined, pd.DataFrame([combined_row])], ignore_index=True)
```

```python
# Voting data is cleaned (Anna)
voting_final = voting_combined
voting_final

# Export cleaned cdf dataframe as csv
voting_final.to_csv('voting_final.csv', index=False)
```
